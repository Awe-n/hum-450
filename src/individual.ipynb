```python
import pandas as pd

# filename = "../xlsx/pension/1875_pension.xlsx"
filename = "../xlsx/pension/1875_pension.xlsx"
df = pd.read_excel(filename)

# Read the data from file "toponyms.xlsx" located in the "./xlsx/toponyms/" folder
toponyms = pd.read_excel("../xlsx/toponyms/toponyms.xlsx")

# Compare the value of the column "pensionnaires_origine" of the dataframe "df" with the values of the column "name" of the dataframe "toponyms"
# If one of the values of the dataframe "toponyms" is found in the value of the column "pensionnaires_origine" of the dataframe "df", then save the value of "penionnaires_origine" in a new dataframe "df_origine"
df_origine = []
for i in range(len(df)):
    for j in range(len(toponyms)):
        ## convert to lower case for comparison
        if toponyms["name"][j].lower() not in df["pensionnaires_origine"][i].lower():
            ## print the values of the toponyms found in the dataframe "df" and of the pensionnaires_origine value
            # print("Toponym found: ", toponyms["name"][j], " in ", df["pensionnaires_origine"][i])
            df_origine.append(df["pensionnaires_origine"][i])
            break

# print the dataframe "df_origine"
print(df_origine)
```

```python
df.head()

# compute the number of pensionnaires for each row.
# to do so, count the occurences of the value "|" in the column "pensionnaires_nom"

# create a new column "pensionnaires_count" in the dataframe a copy of the dataframe "df" called "df_count"
df_count = df.copy()
# create a new column "pensionnaires_count" in the dataframe "df_count" and initialize it with 0
df_count["pensionnaires_count"] = 0

# for each row of the dataframe "df_count"
for i in range(len(df_count)):
    # count the occurences of the value "|" in the column "pensionnaires_nom"
    df_count["pensionnaires_count"][i] = df_count["pensionnaires_nom"][i].count("|") + 1

# print the dataframe "df_count"
df_count.head()
```

```python
# set df pensionnaires to an empty dataframe
df_pensionnaires = pd.DataFrame()
errors = []
valid_indexes = 0

# iterate over the rows of the dataframe "df_count", and for each row, create a number of rows for a new dataframe df_pensionnaires, equal to the value of the column "pensionnaires_count"
for i in range(len(df_count)):
# for i in range(3): 

    fault = False
    # for each row of the dataframe "df_count", get the value of the column "pensionnaires_nom" split by the value "|"
    splitted_pensionnaires_nom = df_count["pensionnaires_nom"][i].split("|")
    # for each row of the dataframe "df_count", get the value of the column "pensionnaires_origine" split by the value "|"
    splitted_pensionnaires_origine = df_count["pensionnaires_origine"][i].split("|")
    # for each row of the dataframe "df_count", get the value of the column "pensionnaires_prenom" split by the value "|"
    splitted_pensionnaires_prenom = df_count["pensionnaires_prenom"][i].split("|")
    # for each row of the dataframe "df_count", get the value of the column "pensionnaires_condition" split by the value "|"
    # check if the value of the column "pensionnaires_condition" is a string
    if type(df_count["pensionnaires_condition"][i]) == str:
        splitted_pensionnaires_condition = df_count["pensionnaires_condition"][i].split("|")
    else:
        errors.append((i, "pensionnaires_condition"))
        fault = True

    # check that the length of all those lists are equal to the value of the column "pensionnaires_count"
    # if not, add the index of the row to the list "errors"
    if len(splitted_pensionnaires_nom) != df_count["pensionnaires_count"][i]:
        errors.append((i, "pensionnaires_nom"))
        fault = True
    if len(splitted_pensionnaires_origine) != df_count["pensionnaires_count"][i]:
        errors.append((i, "pensionnaires_origine"))
        fault = True
    if len(splitted_pensionnaires_prenom) != df_count["pensionnaires_count"][i]:
        errors.append((i, "pensionnaires_prenom"))
        fault = True
    if len(splitted_pensionnaires_condition) != df_count["pensionnaires_count"][i]:
        errors.append((i, "pensionnaires_condition"))
        fault = True

    if not fault:
        # print the number of rows in the dataframe "df_pensionnaires"
        for j in range(df_count["pensionnaires_count"][i]):
            # if the new pensionnaires_condition contains ("pension|pensre"), then set the value of the column "pensionnaires_condition" to "pension"
            if splitted_pensionnaires_condition[j] == "pension" or splitted_pensionnaires_condition[j] == "pensre" or splitted_pensionnaires_condition[j] == "pensionnaire" or splitted_pensionnaires_condition[j] == "pensionnaires":
                splitted_pensionnaires_condition[j] = "pension"
                
            # copy the row of the dataframe "df_count" at index i to the dataframe "df_pensionnaires"
            df_pensionnaires = pd.concat([df_pensionnaires, df_count.iloc[[i]]])             
            valid_indexes += 1
            # set the index of the row of the dataframe "df_pensionnaires" to the value of the variable "valid_indexes"
            df_pensionnaires.index = range(valid_indexes)
            
            df_pensionnaires.loc[df_pensionnaires.index[-1], "pensionnaires_nom"] = splitted_pensionnaires_nom[j]
            df_pensionnaires.loc[df_pensionnaires.index[-1], "pensionnaires_origine"] = splitted_pensionnaires_origine[j]
            df_pensionnaires.loc[df_pensionnaires.index[-1], "pensionnaires_prenom"] = splitted_pensionnaires_prenom[j]
            df_pensionnaires.loc[df_pensionnaires.index[-1], "pensionnaires_condition"] = splitted_pensionnaires_condition[j]

# print the dataframe "df_pensionnaires"
df_pensionnaires.head()
```

```python
# filter the dataframe "df_pensionnaires" to keep only the rows where the value of the column "pensionnaires_condition" is equal to "pension"
df_pensionnaires_pension = df_pensionnaires[df_pensionnaires["pensionnaires_condition"] == "pension"]

# print the length of the dataframe "df_pensionnaires_pension"
print(len(df_pensionnaires_pension))
```

```python
# save the dataframe "df_pensionnaires" to the file "1895_pension_individual.xlsx" located in the "./csv/pension/individual/" folder
df_pensionnaires_pension.to_excel("../xlsx/pension/individual/1875_individual_pension.xlsx")
```
