```python
# import the pandas library
import pandas as pd

# import the data from the files "1855_filtered-full.xlsx" and "1835_v2_filtered-full.xlsx"
df_1855 = pd.read_excel("1855_filtered-full.xlsx")
df_1835 = pd.read_excel("1835_v2_filtered-full.xlsx")

## print the name of the columns of the files "1855_filtered-full.xlsx" and "1835_v2_filtered-full.xlsx"
# print(df_1855.columns)
# print(df_1835.columns)

## compare the value of the column "chef_nom" of the files "1855_filtered-full.xlsx" and "1835_v2_filtered-full.xlsx"
## store the "chef_nom" values of the file "1835_v2_filtered-full.xlsx" in a list
list_chef_nom_1835 = df_1835["chef_nom"].tolist()
## store the "chef_nom" values of the file "1855_filtered-full.xlsx" in a list
list_chef_nom_1855 = df_1855["chef_nom"].tolist()

## print the number of "chef_nom" values of the file "1835_v2_filtered-full.xlsx"
print("Number of chef_nom values in 1835_v2_filtered-full.xlsx: " + str(len(list_chef_nom_1835)))

## print the number of "chef_nom" values of the file "1855_filtered-full.xlsx"
print("Number of chef_nom values in 1855_filtered-full.xlsx: " + str(len(list_chef_nom_1855)))

## iterate over the "chef_nom" values of list_chef_nom_1835 and check if they are in list_chef_nom_1855
## if they are in list_chef_nom_1855, save the value in a list "common_chef_nom"
## also save the index of the value in a list of tuples "common_chef_nom_index"
common_chef_nom = []
common_chef_nom_index = []
for i in range(len(list_chef_nom_1835)):
    if list_chef_nom_1835[i] in list_chef_nom_1855:
        common_chef_nom.append(list_chef_nom_1835[i])
        common_chef_nom_index.append((i, list_chef_nom_1855.index(list_chef_nom_1835[i])))

## print the number of "chef_nom" values that are in both files
print("Number of chef_nom values that are in both files: " + str(len(common_chef_nom)))

## iterate over the tuples in common_chef_nom_index and get the values of the columns "pensionnaires_nom" of the files "1855_filtered-full.xlsx" and "1835_v2_filtered-full.xlsx"
## save the value of the column "pensionnaires_nom" of the file "1835_v2_filtered-full.xlsx" and the value of the column "pensionnaires_nom" of the file "1855_filtered-full.xlsx" in a list of tuples
common_pensionnaires_nom = []
for i in range(len(common_chef_nom_index)):
    common_pensionnaires_nom.append((df_1835["pensionnaires_nom"][common_chef_nom_index[i][0]], df_1855["pensionnaires_nom"][common_chef_nom_index[i][1]]))

## print the number of tuples in common_pensionnaires_nom
print("Number of tuples in common_pensionnaires_nom: " + str(len(common_pensionnaires_nom)))

## iterate over the tuples in common_pensionnaires_nom and print the values of the tuples
for i in range(len(common_pensionnaires_nom)):
    print(common_pensionnaires_nom[i])
```
