```python
import pandas as pd

filename = "./filtered/1895 copy.xlsx"
df = pd.read_excel(filename)

## print the first 5 rows of the data
df.head()
```

```python
## filter the data to only include the rows where the value of the column "pensionnaires_condition" is not empty or NaN or "."
df_condition = df[df["pensionnaires_condition"].notnull() & (df["pensionnaires_condition"] != "·")]

## print the first 5 rows of the filtered data
df_condition.head()
```

```python
## filter the rows where the value of the column "pensionnaires_condition" contains "pension" or "pensre"
df_pensionnaires = df_condition[df_condition["pensionnaires_condition"].str.contains("pension|pensre", na=False)] 

## print the first 5 rows of the filtered data
df_pensionnaires.head()

## print the number of rows in the filtered data
df_pensionnaires.shape
```

```python
## print the values of the column "chef_vocation" of the filtered data
df_pensionnaires["chef_vocation"]
```

```python
## print the values of the column "pensionnaires_origine" of the filtered data
df_pensionnaires["pensionnaires_origine"]
```

```python
## save the filtered data to a new file located in the folder "pension"
df_pensionnaires.to_excel(filename.replace("filtered", "pension"))
```
