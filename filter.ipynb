```python
## open the file 1855.xlsx
## read the first sheet

import pandas as pd

excel_name = '1835_v2.xlsx'

df = pd.read_excel(excel_name, sheet_name=0)

## print the first 5 rows

df.head()
```

```python
## print the column names of the dataframe
df.columns

## save in the variable 'useful_columns' the columns that contain the word 'pensionnaires'
useful_columns = [col for col in df.columns if 'pensionnaires' in col]

## print the useful columns
useful_columns
```

```python
## create a new dataframe with only the useful columns
df2 = df[useful_columns]

## print the first 5 rows of the new dataframe
df2.head()
```

```python
## print the value of the first row for each column
df2.iloc[0]

## print only the value
df2.iloc[0].values

## filter the dataframe to keep only the rows that contain a value in the first column or the second column that is not '·'
df3 = df2[(df2.iloc[:,0] != '·') | (df2.iloc[:,1] != '·')]

## filter the dataframe df2 to keep only the rows that contain a value in the first column or the second column or the third column or the fourth column that is not '·'
df4 = df2[(df2.iloc[:,0] != '·') | (df2.iloc[:,1] != '·') | (df2.iloc[:,2] != '·') | (df2.iloc[:,3] != '·')]
```

```python
## print the first 5 rows of the new dataframe
df4.head()
```

```python
## count the number of rows in the dataframe
df4.shape[0]
```

```python
## export the dataframe to a xlsx file, with the previous name concatenated with '_filtered'
df4.to_excel(excel_name[:-5] + '_filtered.xlsx', index=False)
```
